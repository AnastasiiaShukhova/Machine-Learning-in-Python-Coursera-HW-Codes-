```python
from sklearn.decomposition import PCA
import numpy as np
import pandas 
import scipy
```

Загрузите данные close_prices.csv. В этом файле приведены цены акций 30 компаний на закрытии торгов за каждый день периода.

```python
close = pandas.read_csv('close.csv')
close.head()
```

На загруженных данных обучите преобразование PCA с числом компоненты равным 10. Скольких компонент хватит, чтобы объяснить 90% дисперсии?

```python
pca_close = PCA(n_components=10)
pca_close.fit_transform(close.iloc[:, 1:])
print(pca_close.explained_variance_ratio_) 
np.sum(pca_close.explained_variance_ratio_[:4])
# 4 компоненты

components = pca_close.components_
print components.shape
```

Примените построенное преобразование к исходным данным и возьмите значения первой компоненты.


```python
transformed = pca_close.transform(close.iloc[:, 1:])
first_component = transformed[:, 0]
print first_component.shape
```
